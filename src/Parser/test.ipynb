```python
import socket
import struct
import json

def query_server(text, host='localhost', port=18783):
    try:
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
            s.connect((host, port))
            
            # Prepare the message
            dict_path = "example.PubTator"  # Adjust this if necessary
            s.sendall(dict_path.encode('utf-8'))
            
            # Receive the response length
            raw_length = s.recv(2)
            if len(raw_length) < 2:
                raise ValueError("Did not receive correct message length from server.")
                
            message_length = struct.unpack('>H', raw_length)[0]
            
            # Receive the response
            response = s.recv(message_length)
            if len(response) < message_length:
                raise ValueError("Did not receive the full message from the server.")
            
            result = json.loads(response.decode('utf-8'))
            return result
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
example_text = "BRCA1 and BRCA2 mutations."

# Query the server
result = query_server(example_text)

# Print the results
if result:
    print(json.dumps(result, indent=4))
```
