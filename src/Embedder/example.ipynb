```python
from sentence_transformers import SentenceTransformer

sentences = [
    "The patient's tumor harbors NRAS mutation",
    "Patients with RAS mutations are eligibile for the trial"
]

model = SentenceTransformer('dmlls/all-mpnet-base-v2-negation')
embeddings = model.encode(sentences)
print(embeddings)
```

```python
from scipy import spatial
def cosine_similarity(embeddings):
    return 1 - spatial.distance.cosine(embeddings[0], embeddings[1])
```

```python
cosine_similarity(embeddings)
```
